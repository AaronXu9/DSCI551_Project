```python
# button
button_operation_description = widgets.Button(
                description='Get Operation Description',
                tooltip='Send',
                layout={'width':'auto','margin':'1em auto'}

            )
button_send = widgets.Button(
                description='Get Result',
                tooltip='Send',
                layout={'width':'auto','margin':'1em auto'}   
            )
button_film = widgets.Button(
                description='Search',
                tooltip='Send',
                layout={'width':'auto','margin':'1em auto'}   
            )
button_analysis = widgets.Button(
                description='Analysis',
                tooltip='Send',
                layout={'width':'auto','margin':'auto'}   
            )

output_description = widgets.Output(layout={'border': '1px solid black','width':'450px','margin':'1em auto'})
output_result = widgets.Output(layout={'width':'auto','height':'250px','margin':'1em auto'})
result_box = widgets.VBox([output_result])
output_film = widgets.Output(layout={'width':'auto','margin':'auto','height':'250px'})
output_analysis = widgets.Output(layout={'border': '1px solid black','width':'auto','margin':'1em auto'})


def button_clicked_description(event):
    with output_description:
        clear_output()
        if operations.value == 'mkdir':
            print('Creating a directory in file system \nInput Example: /metadata')
        if operations.value == 'ls':
            print('Listing content of a given directory \nInput Example: /partition')
        if operations.value == 'cat':
            print('Displaying content of a file \nInput Example: /metadata/root/movies_metadata')
        if operations.value == 'rm':
            print('Removing a file from the file system \nInput Example: /metadata/root/movies_metadata')
        if operations.value == 'put':
            print('Uploading a file to file system. You are supposed to enter two inputs splitted by comma: file,partition number \nInput Example: movies_metadata.csv,3')
        if operations.value == 'Get Partition Locations':
            print('Get the locations of partitions of the file \nInput Example: /metadata/root/movies_metadata')
        if operations.value == 'Read Partition':
            print('Get the content of partition number of the specified file.You are supposed to enter two inputs splitted by comma: file directory,partition number \nInput Example: /metadata/root/movies_metadata,1')
            
def button_clicked_result(event):
    with output_result:
        clear_output()
        
        url = 'https://dsci551final-52aac-default-rtdb.firebaseio.com' + variables.value + '/.json'
        if operations.value == 'mkdir':
            try:
                mkdir(url)
                print(f'{variables.value} has been created')
            except:
                print('Plase enter a valid input')
        if operations.value == 'ls':
            try:
                ls(url)
            except:
                print('')
        if operations.value == 'cat':
            pd.set_option("display.max_rows",None)
            display(preprocess_partition_str(cat(url)))
        if operations.value == 'rm':
            try:
                rm(url)
                print(f'{variables.value} has been removed')
            except:
                print('Plase enter a valid input')
        if operations.value == 'put':
            try:
                variable = variables.value.split(',')
                if len(variable) != 2:
                    print('Please enter correct variables!')
                else:
                    file = variable[0]
                    print(file)
                    url_metadata = 'https://dsci551final-52aac-default-rtdb.firebaseio.com/metadata/root/' #need to modify later
                    url_partition = 'https://dsci551final-52aac-default-rtdb.firebaseio.com/partition'
                    k = int(variable[1])
                    print(k)
                    put(file,url_metadata,url_partition,k)
            except:
                print('Plase enter a valid input')
        if operations.value == 'Get Partition Locations':
            try:
                getPartitionLocations(url)
            except:
                print('Plase enter a valid input')
        if operations.value == 'Read Partition':
            try:
                variable = variables.value.split(',')
                if len(variable) != 2:
                    print('Please enter correct variables!')
                else:
                    url_read = 'https://dsci551final-52aac-default-rtdb.firebaseio.com' + variable[0] + '/.json'
                    partition_num = variable[1]
                    partition_list = readPartition(url_read,partition_num)
                    pd.set_option("display.max_rows",None)
                    display(preprocess_partition_str(partition_list))
            except:
                print('Plase enter a valid input')

def button_clicked_film(event):
    with output_film:
        clear_output()
        try:
            rootURL = 'https://dsci551final-52aac-default-rtdb.firebaseio.com/'
            language = filter_language.value

            if filter_rating.value:
                rating = float(filter_rating.value)
            else:
                rating = filter_rating.value
            genre = filter_genre.value
            s_date = filter_s_date.value
            e_date = filter_e_date.value
            if s_date == None:
                s_date = '1900-01-01'
            else:
                s_date = filter_s_date.value.strftime("%Y-%m-%d")
            if e_date == None:
                e_date = '2022-11-01'
            else:
                e_date = filter_e_date.value.strftime("%Y-%m-%d")
                
            df = pmr(rootURL=rootURL, language=language, rating=rating, genre=genre, s_date=s_date, e_date=e_date)
            df = df.drop(['id','production_countries'],axis=1)
            pd.set_option("display.max_rows",None)
            display(df)
        except:
            print('Plase enter a valid input')

def button_clicked_analysis(event):
    with output_analysis:
        clear_output()
        try:
            rootURL = 'https://dsci551final-52aac-default-rtdb.firebaseio.com/'
            language = filter_language.value

            if filter_rating.value:
                rating = float(filter_rating.value)
            else:
                rating = filter_rating.value
            genre = filter_genre.value
            s_date = filter_s_date.value
            e_date = filter_e_date.value
            if s_date == None:
                s_date = '1900-01-01'
            else:
                s_date = filter_s_date.value.strftime("%Y-%m-%d")
            if e_date == None:
                e_date = '2022-11-01'
            else:
                e_date = filter_e_date.value.strftime("%Y-%m-%d")
                
            df = pmr(rootURL=rootURL, language=language, rating=rating, genre=genre, s_date=s_date, e_date=e_date)
            df = df.drop(['id','production_countries'],axis=1)
            avg, max_film, min_film = analyze(df)
            print(f'The average rating of the films are: {avg}\n')
            print(f'The Highest Rated Film(s):\n')
            pd.set_option("display.max_rows",None)
            display(max_film)
            print(f'The Lowest Rated Film(s):\n')
            display(min_film)
        except:
            print('No Results Available')
            
button_send.on_click(button_clicked_result)
button_operation_description.on_click(button_clicked_description)
button_film.on_click(button_clicked_film)
button_analysis.on_click(button_clicked_analysis)
```
